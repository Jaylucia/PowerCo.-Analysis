**Distributions of Numerical Columns (Customer data)**

```python
client_df.hist(figsize=(14, 10), bins=30)
plt.suptitle("Customer Data - Distributions")
plt.tight_layout()
plt.show()
```

The distribution of key customer consumption variables reveals a consistent right-skewed pattern across multiple metrics, including electricity consumption over the last 12 months, gas consumption over the same period, consumption in the last month, as well as forecasted electricity consumption and forecasted discount energy for the next 12 months.

This right skew indicates that the majority of customers are concentrated around lower to moderate levels of energy usage, as evidenced by the higher frequency of observations on the left side of the distributions. The elongated tail on the right reflects a smaller group of customers with significantly higher consumption levels.

Although the overall pattern remains consistent, the distribution of forecasted electricity consumption over the next 12 months appears more dispersed. This suggests that while most customers are still expected to maintain moderate usage levels, there exists a subset of customers projected to consume substantially higher amounts of electricity in the future. A similar pattern is observed in the forecasted bill for meter rental, current paid consumption, and gross margin on power subscription, where the spread of values indicates the presence of high-value or high-usage customers within the dataset.

For the forecasted energy price variables, more nuanced patterns emerge. The forecasted energy price for the second period retains a right-skewed structure but exhibits a concentration of values around the center, suggesting partial symmetry and a more stable pricing range for most customers.

In contrast, the forecasted energy price for the first period, particularly for off-peak consumption, displays an inverse pattern, indicating a distribution where higher values are more common and lower values are less frequent. This may reflect differences in pricing structures or demand patterns between peak and off-peak periods.

The distribution of customer tenure, measured by the number of years a customer has been with the company, is strongly concentrated between 3 and 6 years, with a noticeable peak around 5 to 6 years. Beyond year 7, the number of customers declines sharply, and very few customers remain beyond the 10-year mark.

This pattern suggests that customer retention weakens significantly after the mid-tenure period. The sharp drop in customer count beyond 7 years may indicate a higher likelihood of churn occurring between 6 and 8 years of tenure. ***This is a critical insight that warrants further investigation when analyzing the churn variable***, as it may highlight a key window where targeted retention strategies could be most effective.

**Analyzing the Relationship between Churn and Key Variables**

```python
client_df.groupby('churn').mean(numeric_only=True)
```

```python
cols = ['cons_12m', 'cons_gas_12m', 'net_margin', 'num_years_antig']
client_df.groupby('churn')[cols].mean()
```

The comparison between churned and retained customers reveals clear and meaningful differences in consumption behavior, profitability, and tenure.

Customers who churned consistently exhibited significantly lower energy consumption across both electricity and gas metrics. In particular, churned customers consumed less than half the electricity and approximately one-third of the gas compared to retained customers. This pattern aligns with expected business behavior, as customers with lower usage levels are typically less dependent on the service and therefore face lower switching costs. As a result, they are more likely to explore alternative providers or discontinue the service altogether.

Despite their lower consumption levels, churned customers generated higher net margins on average, with a mean margin of 228 compared to 185 for retained customers. This represents a particularly important and somewhat counterintuitive finding. It suggests that these customers were either subject to relatively higher pricing structures or were associated with contract types that yielded stronger margins for the company. From a business perspective, this implies that the loss of these customers has a disproportionately negative impact on profitability, as the company is losing customers who contribute more profit per unit of consumption.

In terms of customer tenure, churned customers tend to have slightly shorter relationships with the company. This indicates that they are relatively newer customers who may not have fully integrated into the service. When considered alongside their lower consumption patterns, this suggests a segment of customers who never developed strong engagement or reliance on the service, making them more susceptible to churn.

**Key Insight:**

Churned customers represent a distinct segment characterized by low consumption but high profitability. Although they contribute less in terms of total energy usage, they generate higher margins on a per-customer basis. This indicates that the company is not simply losing low-value customers, but rather a financially important segment that contributes disproportionately to profit.

This pattern suggests that these customers may be switching to competitors offering more attractive pricing structures tailored to low-usage profiles. As such, targeted retention strategies focused on pricing optimization or tailored plans for low-consumption customers could help mitigate churn and protect a valuable revenue stream.

**Distributions of Numerical Columns (Price data)**

```python
price_df.hist(figsize=(14, 10), bins=30)
plt.suptitle("Price Data - Distributions")
plt.tight_layout()
plt.show()
```

The analysis of pricing variables highlights clear structural patterns in how customers are charged across different periods, with important implications for customer behavior and potential churn dynamics.

Off-peak variable pricing follows a near-normal distribution, indicating that the majority of customers are charged similar rates within a relatively narrow range. This suggests a largely standardized pricing structure for off-peak usage, with limited variation across the customer base. Only a small number of customers fall outside this central range, paying either slightly higher or lower rates.

In contrast, both peak and mid-peak variable pricing exhibit right-skewed distributions. Most customers are clustered around moderate pricing levels, while a smaller segment is exposed to significantly higher rates. This dispersion likely reflects differences in contract types, where certain customers may be on older, less competitive plans or premium contracts that result in elevated peak-time charges.

Off-peak fixed pricing presents a bimodal distribution, indicating the presence of two distinct pricing clusters within the customer base. This pattern strongly suggests the existence of two separate contract tiers or pricing structures, where one group of customers is consistently charged lower fixed rates while another group is subject to higher fixed charges, with relatively few customers positioned between these extremes.

```python
plt.figure(figsize=(8, 4))
sns.histplot(price_df['price_off_peak_fix'], bins=30, kde=True)
plt.title("Distribution of Off-Peak Fixed Price")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

The pricing dataset reflects two fundamental components of customer charges: variable prices, which depend on actual energy consumption, and fixed prices, which represent standing charges applied regardless of usage. These prices are structured across three time periods: off-peak, peak, and mid-peak, each capturing different demand conditions and pricing dynamics.

Across the variable pricing structure, off-peak energy prices display a near-normal distribution, indicating that most customers are charged within a narrow and consistent range. This concentration suggests a high level of standardization in off-peak pricing, with limited differentiation between customers. In contrast, both peak and mid-peak variable prices exhibit right-skewed distributions. While the majority of customers pay moderate rates, a smaller segment faces significantly higher prices. This pattern points to variation in contract structures, where certain customers may be on legacy or premium plans that result in elevated charges during higher-demand periods.

A similar pattern emerges within fixed pricing for peak and mid-peak periods. Both variables are right-skewed, with most customers clustered around moderate fixed charges and a smaller group subject to higher rates. This reinforces the presence of differentiated pricing structures across the customer base, likely tied to contract type or pricing plans.

Off-peak fixed pricing, however, stands out due to its bimodal distribution. The presence of two distinct clusters indicates that customers are segmented into two clearly separated pricing tiers. One group is consistently charged lower fixed rates, while the majority falls within a higher pricing band, with minimal overlap between the two. This structure strongly suggests the coexistence of legacy contracts alongside newer or standard pricing schemes.

From a business perspective, this disparity introduces a potential source of perceived inequity. Customers on higher fixed rates may become dissatisfied if they recognize that others are paying significantly less for comparable service. As a result, this pricing structure may contribute to churn, particularly among customers exposed to less favorable terms.

Overall, the pricing data suggests that while variable pricing is relatively standardized, fixed pricing structures, especially in the off-peak period, reveal meaningful segmentation within the customer base. This variation is likely to play an important role in shaping customer behavior and should be examined further as a potential driver of churn in subsequent analysis stages.

4. The company loses its most profitable customers. Net margin

The comparison of pricing variables between churned and retained customers highlights where pricing may, and may not, influence customer decisions.

Off-peak variable pricing shows no meaningful difference between the two groups, with both churned and retained customers paying approximately $0.04 per unit. This consistency indicates that off-peak variable pricing is highly standardized and does not appear to play a role in churn behavior.

Peak variable pricing, however, reveals a small but relevant difference. Churned customers paid approximately $0.02 per unit compared to $0.01 for retained customers. While the absolute difference is modest, peak periods typically account for a significant portion of total energy consumption. As a result, even a slight increase in unit cost can accumulate into a noticeable difference in annual billing, potentially influencing customer decisions over time.

Off-peak fixed pricing shows only a marginal difference, with churned customers paying $13.02 compared to $12.08 for retained customers. This gap is relatively small and, in isolation, is unlikely to be a primary driver of churn.

Peak fixed pricing stands out as the most significant finding in the pricing analysis. Churned customers paid $3.56 compared to $2.68 for retained customers, representing an increase of approximately 33 percent. This is a substantial difference in standing charges and is likely to have a direct impact on customer perception of value. Unlike variable pricing, fixed charges are highly visible and consistent on customer bills, making them more salient. This pricing gap strongly suggests that higher peak fixed charges may have contributed to customer dissatisfaction and increased the likelihood of switching to alternative providers.
